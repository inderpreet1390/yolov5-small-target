```python
import os
import glob
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
path=r'I:\ISR2\Datasets labels\Dota\all_1644706476.txt'
```

```python
file = open(path,"r")
Counter = 0

Content = file.read()
CoList = Content.split("\n")

fls=[]
  
for i in CoList:
    if i:
        ls=i.split()
        ar=(float(ls[2])-float(ls[0]))*(float(ls[5])-float(ls[3]))
        Counter += 1
        fls.append(ar)
          
print("This is the number of lines in the file")
print(Counter)
```

```python
fls
```

```python
len(fls)
```

```python
figure(num=1,figsize=(14, 10))
plt.hist(fls, 20000)
plt.ylim([0, 50000])
plt.xlim([0, 10000])
plt.xlabel("Area in pixels squared")
plt.ylabel("Number of instances")
plt.title("Histogram of Area of objects in pixel squared for DOTA dataset",y=-0.1)
plt.axvline(x=1024,color='red',label='1024')
plt.axvspan(0,1024,color='red',label="<1024",alpha=0.3)
plt.legend()
plt.show()
```

```python
path2=r'I:\ISR2\Datasets labels\Visdrone\all_1644708653.txt'
```

```python
file2 = open(path2,"r")
Counter2 = 0

Content2 = file2.read()
CoList2 = Content2.split("\n")

fls2=[]
  
for i in CoList2:
    if i:
        ls=i.split(',')
        ar=(float(ls[2])*float(ls[3]))
        Counter2 += 1
        fls2.append(ar)
          
print("This is the number of lines in the file")
print(Counter2)
```

```python
fls2
```

```python
len(fls2)
```

```python
figure(num=1,figsize=(14, 10))
plt.hist(fls2, 20000)
plt.ylim([0, 8000])
plt.xlim([0, 5000])
plt.xlabel("Area in pixels squared")
plt.ylabel("Number of instances")
plt.title("Histogram of Area of objects in pixel squared for VisDrone dataset",y=-0.1)
plt.axvline(x=1024,color='red',label='1024')
plt.axvspan(0,1024,color='red',label="<1024",alpha=0.3)
plt.legend()
plt.show()
```

```python
cfls=0
for i in fls:
    if i<1024:
        cfls=cfls+1
print("Percentage of labels that are considered small as per COCO evaluation(smaller than 1024 pixel area): ",(cfls/len(fls))*100, " %")
```
